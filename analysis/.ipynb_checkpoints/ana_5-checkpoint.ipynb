## Analyzing Map of Chicago with maximum Crime Occurrences in past 2 Years

```python
#Import the required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
from matplotlib import cm
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
```

```python
#Read the input file
df=pd.read_csv('../data/raw_data/Crimes_-_2001_to_present.csv')
```

```python
#Print the input
df.head()
```

```python
#Limiting data for past 2 years
df = df[df['Year'] > 2014]
```

```python
#Lets try to plot Crime on map based on Geolocation
east = df['Longitude'].max()
west = df['Longitude'].min()
north = df['Latitude'].max()
south = df['Latitude'].min()
#lon = df['Longitude'].mean()
#lat = df['Latitude'].mean()

fig,ax = plt.subplots(figsize = (10,10))
ax.set_title("Chicago Crime Heat Map in Last 2 Years")

m = Basemap(projection='merc', llcrnrlat=south, urcrnrlat=north,
            llcrnrlon=west, urcrnrlon=east, lat_ts=south, resolution='h')

m.drawmapboundary(fill_color='#46bcec')
#m.fillcontinents(lake_color='aqua')
m.drawcoastlines()

x, y = m(df['Longitude'].values, df['Latitude'].values)
m.hexbin(x, y, gridsize=250, bins='log', cmap='inferno');
if not os.path.exists('ana_5'):
    os.makedirs('ana_5')
plt.savefig('ana_5/Chicago Crime Heat Map in Last 2 Years')
```

### Inference
- The Bright spots show more crime activity in past 2 years and light spots show less crime activity in past 2 years.
- Brighter the spot more is it unsafe to stay in that locality.
- If someone is considering moving to Chicago, they should try to live near places with light spots so as to avoid Crime activity.
